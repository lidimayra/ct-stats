```python
arquivo = 'pesos.txt'
```

```python
conteudo = open(arquivo)
```

```python
dados = conteudo.read()
```

```python
conteudo.close()
```

```python
print(dados)
```

```python
lista = str.split(dados)
```

```python
print(lista)
```

```python
inteiros = []
```

```python
for elemento in lista:
    inteiros.append(int(elemento))
```

```python
print(inteiros)
```

```python
universo_amostral = len(inteiros)
```

```python
print(universo_amostral)
```

```python
print(inteiros)
```

```python
frequencias = {}
for xi in inteiros:
    if xi not in frequencias:
        frequencias[xi] = 1
    else:
        frequencias[xi] += 1
```

```python
print(frequencias)
```

```python
import collections
frequencias = collections.OrderedDict(sorted(frequencias.items()))
print(frequencias)
```

```python
xi = list(frequencias.keys())
print(xi)
```

```python
fi = list(frequencias.values())
print(fi)
```

```python
%matplotlib notebook

import pylab
pylab.figure(1)
pylab.plot(xi, fi, 'r--')
pylab.title("Pesos")
pylab.xlabel("xi")
pylab.ylabel("fi")
pylab.show()
```
